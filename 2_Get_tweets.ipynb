```python
import tweepy
import numpy as np
import json
import pickle
import csv
from twitter import Twitter, OAuth, TwitterHTTPError
from time import time

# get my Twitter API credentials
from twitter_api_keys import mykeys
```

```python
auth = tweepy.OAuthHandler(mykeys['CONSUMER_KEY'], mykeys['CONSUMER_SECRET'])
auth.set_access_token(mykeys['ACCESS_TOKEN'], mykeys['ACCESS_SECRET'])
api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
users = np.load(open("usernames_list.npy", "rb" ))
```

```python
users = list(users)
```

```python
len(users)
```

```python
def get_all_tweets(users): 
        
        
    # make request for most recent tweets (200 is the maximum allowed count)
    for user in users:
        try:
            new_tweets = api.user_timeline(screen_name = user, count=200, trim_user=True, )

            #save desired fields from the user's 200 tweets
            outtweets = [[str(user), str(tweet.id_str), str(tweet.text)] for tweet in new_tweets]

            # append the user's tweets to a csv
            with open('tweets.csv', 'a', newline='') as f:
                writer = csv.writer(f)
                writer.writerows(outtweets)
                f.close()
            print('saved {} tweets for user {} to csv'.format(len(new_tweets), user))
                
        except:
            print('passing')
            pass
```

```python
alltweets = get_all_tweets(users)
```
